```python
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
import os
import numpy as np
from PIL import Image
import glob
```

```python
def createfolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
            pass
```

```python
images=[glob.glob('./Data_Nrom/*.jpg'),glob.glob('./Data_MOT/*.jpg'),glob.glob('./Data_AbNorm/*.jpg')]
images=np.array(images).tolist()
```

```python
def make_my_picutre(images,root,data_type):
    for i in range(len(images)):
        image=cv2.imread(images[i])
        image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        IMAGE_SIZE= 340
        train_transform = A.Compose([
            #A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=1.0)
            A.Resize(300, 300), 
            A.RandomCrop(224, 224),
            #A.HorizontalFlip(p=0.5)
            #A.VerticalFlip(p=0.5)
            #A.RandomRotate90(p=1)
            #A.ShiftScaleRotate(shift_limit=0.4, scale_limit=(0.5, 0.9), rotate_limit=90, p=1, border_mode=cv2.BORDER_REPLICATE)
            #A.Cutout(p=1, num_holes=8, max_h_size=24, max_w_size=24)
            #A.Blur(p=1, blur_limit=(50, 60))
        ])
        transformed=train_transform(image=image)
        transformed_image=transformed['image']
            
        transformed_image=cv2.cvtColor(transformed_image, cv2.COLOR_RGB2BGR)

        createfolder(root)
        cv2.imwrite(f"./{data_type}/{data_type}_{i+1}.png",transformed_image)
```

```python
root_data_type=[['./Mot_aug','Mot_aug'],
               ['./Norm_aug','Norm_aug'],
               ['./AbNorm_aug','AbNorm_aug']]
```
